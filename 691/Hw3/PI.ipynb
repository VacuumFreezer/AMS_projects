```python
import numpy as np
import gymnasium as gym
from util import transition_kernel, policy_iteration, plot_pi_convergence, plot_value_grid, plot_policy_arrows
```

```python
env = gym.make(
    'FrozenLake-v1',
    desc=None,
    map_name="4x4",
    is_slippery=True,
)
desc = env.unwrapped.desc
P, nS, nA = transition_kernel(env)

gamma = 0.9
V_star, pi_star, Q_star, pi_deltas = policy_iteration(P, nS, nA, gamma=gamma, theta_eval=1e-6, max_eval_iters=100)

plot_pi_convergence(pi_deltas[1:])
plot_value_grid(V_star, desc, title='Optimal Value Function')
plot_policy_arrows(pi_star, desc, title='Optimal Policy')

np.set_printoptions(precision=4, suppress=True)
print("Optimal Value Function:\n", V_star.reshape(desc.shape))
print("\nOptimal policy (0:L,1:D,2:R,3:U):\n", pi_star.reshape(desc.shape))
print("\nOptimal Q* (rows=states, cols=actions [L,D,R,U]):\n", Q_star)
```
