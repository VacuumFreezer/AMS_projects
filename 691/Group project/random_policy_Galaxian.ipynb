```python
import gymnasium as gym
import ale_py
import numpy as np

# Register ALE environments
gym.register_envs(ale_py)

def test_random_agent(env_name="ALE/Galaxian-v5", steps=1024):
    print(f"Testing Random Agent on {env_name} for {steps} steps...")
    
    env = gym.make(env_name)
    obs, info = env.reset()
    
    total_reward = 0
    episodes = 0
    
    for _ in range(steps):
        # Select a random action from the action space
        action = env.action_space.sample()
        
        # Execute action
        obs, reward, terminated, truncated, info = env.step(action)
        
        total_reward += reward
        
        # If the episode ends (game over), reset the environment
        if terminated or truncated:
            episodes += 1
            obs, info = env.reset()
            
    env.close()
    
    print("-" * 30)
    print(f"Total Steps: {steps}")
    print(f"Total Reward (Batch Score): {total_reward}")
    print(f"Episodes Completed: {episodes}")
    print("-" * 30)

if __name__ == "__main__":
    test_random_agent()
```
