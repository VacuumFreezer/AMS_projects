```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Decorator to ensure double precision
def double_dec(func):
    def wrapper(*args):
        args = tuple(np.double(arg) for arg in args)

        results = func(*args)

        return tuple(np.double(result) for result in results)
    return wrapper
```

```python
@double_dec
def gen(x_0, x_1, n):
    seq = [x_0, x_1]
    i = 2
    
    x_km1 = x_0
    x_k = x_1
    while i < n:
        x_kp1 = 111.0 - (1130.0-(3000.0/x_km1)) / x_k
        seq.append(x_kp1)
        i += 1
        x_km1 = x_k
        x_k = x_kp1

    return seq
```

```python
seq = gen(11/2, 61/11, 20)
k_value = range(0, 20, 1)

plt.figure(figsize=(8,6))
plt.plot(k_value, seq, linestyle='-', color='blue', label='double')
plt.xlabel('k')
plt.ylabel('seq')
plt.title('Generated sequence')
plt.show()
```
