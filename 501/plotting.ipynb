```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
from math import exp
from math import sqrt
from math import e
```

```python
def y_uniform(x, epsilon):

    y_out = exp(1/(1+x) - 1/2) + epsilon * (1 / (5* (1+x)**5) + 1 / (2*(1+x)**4) - 3.0/80.0) * exp(1/(1+x) - 1/2)

    A_0 = 1 - sqrt(e)
    A_1 = 1 - 133.0/80.0 * sqrt(e)
    cap_X = x / epsilon

    y_in_0 = 1 + A_0 * (exp(-cap_X) - 1)
    y_in_1 = - cap_X**2 * exp(-cap_X) - A_0* cap_X* exp(-cap_X) - A_0* exp(-cap_X) + (A_0-1)* cap_X + 1 - A_0 + (2*A_0-1-A_1) + A_1*exp(-cap_X)
    y_in = y_in_0 + epsilon * y_in_1
    
    y_match = sqrt(e) - sqrt(e)*x + 53.0/80.0 * sqrt(e)*epsilon

    return y_out + y_in - y_match
```

```python
x_data = np.linspace(0, 1, 200)
y_data = np.zeros(len(x_data))

for i in range(len(y_data)):
    y_data[i] = y_uniform(x_data[i], epsilon=0.1)
```

```python
num_data = np.loadtxt('numer_solution_eps0_1.txt')
num_x = num_data[:, 0]
num_y = num_data[:, 1]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, label='perturbation')
plt.plot(num_x, num_y, label='numerical')
plt.title(f'$\epsilon = 0.1$')
plt.legend()
plt.show()
```
